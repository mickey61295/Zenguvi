```python
import pandas as pd
df5m = pd.read_csv(r"C:\Users\mahes\Downloads\OANDA_EURUSD, 5.csv")
df1m = pd.read_csv(r"C:\Users\mahes\Downloads\OANDA_EURUSD, 1.csv")
df5m['signal'] = ''
putlist = []
for i in range(2,len(df5m)-1):
    if df5m['Aroon Up'][i] < df5m['Aroon Up'][i-1] and df5m['Aroon Up'][i-1]>=df5m['Aroon Up'][i-2]:
        df5m['signal'][i+1] = 'Put'
        putlist.append(df5m.time[i+1])

```

```python
df1m['signal'] = ''
pnl = []
co_loss = 0
loss_date = ''
ban_date = ''
bal = 10
initial_bal = 10
in_amt = 1
for i in range(len(df1m)):
    if df1m.time[i] != ban_date:
        if int(df1m.time[i][11:13]) in range(8,23):
            if df1m.time[i] in putlist: # and df1m['Aroon Down'][i] > 60:
                if df1m.open[i] > df1m.close[i]:
                    pnl.append('p')
                    df1m.signal[i] = 'profit'
                    ban_date = ''
                    bal = round((bal-(in_amt+co_loss)) + ((in_amt+co_loss) * 1.8),2)
                    co_loss = 0
                    print(df1m.time[i],bal)
                elif df1m.open[i] < df1m.close[i]:
                    loss_date = df1m.time[i]
                    bal -= (in_amt+co_loss)
                    co_loss += 1
                    if co_loss >= 2:
                        ban_date = df1m.time[i]
                        co_loss = 0
                    pnl.append('l')
                    df1m.signal[i] = 'loss'
                    print(df1m.time[i],bal)
                else:
                    pnl.append('r')
                    df1m.signal[i] = 'return'
            
# print(pnl.count('p'))
# print(pnl.count('l'))
# print(pnl.count('r'))
```
